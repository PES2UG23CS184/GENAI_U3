```python
import torch

d, r = 64, 4
W = torch.randn(d, d)
A = torch.randn(r, d)
B = torch.zeros(d, r)

x = torch.randn(1, d)

output = x @ W.T + (x @ A.T @ B.T)

print("LoRA output shape:", output.shape)
```

```python
import numpy as np

x = np.random.randn(10)

scale = (x.max() - x.min()) / 255
q = np.round(x / scale).astype(np.int8)
x_recon = q * scale

print("Original:", x[:3])
print("Reconstructed:", x_recon[:3])
```
