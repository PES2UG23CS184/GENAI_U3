```python
experts = {
    "math": lambda q: "Math expert answer",
    "code": lambda q: "Code expert answer",
    "general": lambda q: "General answer"
}
```

```python
def router(query):
    if "code" in query:
        return "code"
    elif "math" in query:
        return "math"
    return "general"
```

```python
def moe(query):
    expert = router(query)
    return experts[expert](query)

print(moe("solve math problem"))
```
