```python
from sentence_transformers import CrossEncoder

model = CrossEncoder('cross-encoder/ms-marco-MiniLM-L-6-v2')

# Define 'query' and 'documents' for demonstration purposes
query = "What is the capital of France?"
documents = [
    "Paris is the capital and most populous city of France.",
    "The Eiffel Tower is in Paris.",
    "France is a country in Western Europe."
]

pairs = [(query, doc) for doc in documents]
scores = model.predict(pairs)

for doc, score in zip(documents, scores):
    print(doc, score)
```

```python
from sentence_transformers import SentenceTransformer

# Initialize a SentenceTransformer model for generating embeddings
embedding_model = SentenceTransformer('all-MiniLM-L6-v2')

hypothetical_doc = "Attention mechanisms allow models to focus on important words."

# Generate embedding for the hypothetical document
query_embedding = embedding_model.encode([hypothetical_doc])
print(f"HyDE embedding generated with shape: {query_embedding.shape}")
print(f"First 5 dimensions of the embedding: {query_embedding[0][:5]}")
```
